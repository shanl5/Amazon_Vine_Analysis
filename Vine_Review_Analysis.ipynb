```python
# Open csv file into DataFrame
reviews_df = pd.read_csv("Resources/vine_table.csv")
```

```python
# D2-1.
select_reviews_df = reviews_df[reviews_df["total_votes"] >= 20]

select_reviews_df.head(20)
```

```python
# D2-2.
helpful_reviews_df = select_reviews_df[
    (select_reviews_df["helpful_votes"] / select_reviews_df["total_votes"]) >= 0.5 ]

helpful_reviews_df.head(20)
```

```python
helpful_reviews_df.describe
```

```python
# D2-3.
vine_helpful_df = helpful_reviews_df[helpful_reviews_df["vine"] == "Y"]

vine_helpful_df.head(20)
```

```python
# D2-4.
unpaid_helpful_df = helpful_reviews_df[helpful_reviews_df["vine"] == "N"]

unpaid_helpful_df.head(20)
```

```python
test_df = vine_helpful_df[vine_helpful_df["verified_purchase"] == "Y"]

test_df.head(20)
```

```python
all_test_df = reviews_df[(reviews_df["vine"] == "Y") & (reviews_df["verified_purchase"] == "Y")]

all_test_df.head(20)
```

```python
helpful_reviews_df.count()
```

```python
# D2-5. (Selected Helpful Review Data)

# ------ Determine review totals and percentages
total_helpful_reviews = helpful_reviews_df["review_id"].count()
total_helpful_vine_reviews = vine_helpful_df["review_id"].count()
total_helpful_unpaid_reviews = unpaid_helpful_df["review_id"].count()

percent_helpful_that_were_vine = 100 * (total_helpful_vine_reviews / total_helpful_reviews)
percent_helpful_that_were_unpaid = 100 * (total_helpful_unpaid_reviews / total_helpful_reviews)

# ------ Determine (helpful) five-star review totals and percentages
five_star_helpful_reviews = helpful_reviews_df[helpful_reviews_df["star_rating"] == 5]["review_id"].count()
five_star_vine_reviews = vine_helpful_df[vine_helpful_df["star_rating"] == 5]["review_id"].count()
five_star_unpaid_reviews = unpaid_helpful_df[unpaid_helpful_df["star_rating"] == 5]["review_id"].count()

percent_five_star_that_were_vine = 100 * (five_star_vine_reviews / five_star_helpful_reviews)
percent_five_star_that_were_unpaid = 100 * (five_star_unpaid_reviews / five_star_helpful_reviews)

percent_vine_that_were_5star = 100 * (five_star_vine_reviews / total_helpful_vine_reviews)
percent_unpaid_that_were_5star = 100 * (five_star_unpaid_reviews / total_helpful_unpaid_reviews)

percent_helpful_that_were_5star = 100 * (five_star_helpful_reviews / total_helpful_reviews)

# ======
print(f"Of {total_helpful_reviews:,} reviews noted (D2-2) as being helpful:\n" +
      f"{total_helpful_vine_reviews:,} ({percent_helpful_that_were_vine:.2f}%) were paid 'Vine' reviews, the others " +
      f"({total_helpful_unpaid_reviews:,}; {percent_helpful_that_were_unpaid:.2f}%) being unpaid.")

# print(f"\n")

print(f"\nFor the {five_star_helpful_reviews:,} 5-star helpful reviews considered:\n" +
      f"{five_star_vine_reviews:,} ({percent_five_star_that_were_vine:.2f}%) were 'Vine' reviews; " +
      f"{five_star_unpaid_reviews:,} ({percent_five_star_that_were_unpaid:.2f}%) unpaid.")

print(f"\n\t       Summary of Selected Helpful Reviews" +
      f"\n================================================================\n" +
      f"\t\t\tTotal Number\tNumber 5-Star\t% 5-Star\n" +
      f"\t\t\t------------\t-------------\t--------\n" +
      f"Paid ('Vine') Reviews\t\t{total_helpful_vine_reviews:,}\t\t {five_star_vine_reviews:,}\t  {percent_vine_that_were_5star:.2f}\n" +
      f"Unpaid Reviews\t\t     {total_helpful_unpaid_reviews:,}\t      {five_star_unpaid_reviews:,}\t  {percent_unpaid_that_were_5star:.2f}\n" +
      f"\t\t\t     ------\t      ------\n" +
      f"\tGrand Total\t     {total_helpful_reviews:,}\t      {five_star_helpful_reviews:,}\t  {percent_helpful_that_were_5star:.2f}")
```

```python
# (Full review data)

# ------ Determine overall review totals and percentages
total_overall_reviews = reviews_df["review_id"].count()
total_overall_vine_reviews = reviews_df[reviews_df["vine"] == "Y"]["review_id"].count()
total_overall_unpaid_reviews = reviews_df[reviews_df["vine"] == "N"]["review_id"].count()

percent_overall_that_were_vine = 100 * (total_overall_vine_reviews / total_overall_reviews)
percent_overall_that_were_unpaid = 100 * (total_overall_unpaid_reviews / total_overall_reviews)

# ------ Determine overall five-star review totals and percentages
five_star_overall_reviews = reviews_df[reviews_df["star_rating"] == 5]["review_id"].count()
five_star_overall_vine_reviews = reviews_df[(reviews_df["star_rating"] == 5) & (reviews_df["vine"] == "Y")]["review_id"].count()
five_star_overall_unpaid_reviews = reviews_df[(reviews_df["star_rating"] == 5) & (reviews_df["vine"] == "N")]["review_id"].count()

percent_five_star_overall_that_were_vine = 100 * (five_star_overall_vine_reviews / five_star_overall_reviews)
percent_five_star_overall_that_were_unpaid = 100 * (five_star_overall_unpaid_reviews / five_star_overall_reviews)

percent_overall_vine_that_were_5star = 100 * (five_star_overall_vine_reviews / total_overall_vine_reviews)
percent_overall_unpaid_that_were_5star = 100 * (five_star_overall_unpaid_reviews / total_overall_unpaid_reviews)

percent_overall_that_were_5star = 100 * (five_star_overall_reviews / total_overall_reviews)

# ======
print(f"Of {total_overall_reviews:,} reviews in the data set:\n" +
      f"{total_overall_vine_reviews:,} ({percent_overall_that_were_vine:.2f}%) were paid 'Vine' reviews, the others " +
      f"({total_overall_unpaid_reviews:,}; {percent_overall_that_were_unpaid:.2f}%) being unpaid.")

# print(f"\n")

print(f"\nFor the {five_star_overall_reviews:,} 5-star reviews:\n" +
      f"{five_star_overall_vine_reviews:,} ({percent_five_star_overall_that_were_vine:.2f}%) were 'Vine' reviews; " +
      f"{five_star_overall_unpaid_reviews:,} ({percent_five_star_overall_that_were_unpaid:.2f}%) unpaid.")

print(f"\n\t             Data Set Review Summary" +
      f"\n================================================================\n" +
      f"\t\t\tTotal Number\tNumber 5-Star\t% 5-Star\n" +
      f"\t\t\t------------\t-------------\t--------\n" +
      f"Paid ('Vine') Reviews\t       {total_overall_vine_reviews:,}\t\t  {five_star_overall_vine_reviews:,}\t  {percent_overall_vine_that_were_5star:.2f}\n" +
      f"Unpaid Reviews\t\t     {total_overall_unpaid_reviews:,}\t      {five_star_overall_unpaid_reviews:,}\t  {percent_overall_unpaid_that_were_5star:.2f}\n" +
      f"\t\t\t     -------\t      -------\n" +
      f"\tGrand Total\t     {total_overall_reviews:,}\t      {five_star_overall_reviews:,}\t  {percent_overall_that_were_5star:.2f}")
```
